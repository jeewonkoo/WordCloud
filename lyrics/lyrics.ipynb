```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_BE = pd.read_csv("archive/csv/BillieEilish.csv", index_col=0)
```

```python
df_BE = df_BE.astype({"Lyric": str})
```

```python
lyrics_BE = df_BE["Lyric"]
text_BE = " ".join(lyric for lyric in lyrics_BE)
text_BE = text_BE.upper()
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["YEAH", "LA", "OH", "BOY"])

wordcloud = WordCloud().generate(text_BE)

# Display the generated image:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text_BE)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
