```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re
from matplotlib.colors import LinearSegmentedColormap

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("trump_insult_tweets_2014_to_2021.csv", index_col=0)
#df["tweet"]
```

```python
tweet = df["tweet"]
tweet = tweet.astype(str)
text = " ".join(content for content in tweet)
#Removing URL
text = re.sub(r'^http?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE)

#ADD Stopwords
stopwords.add('say')
stopwords.add('many')
stopwords.add('don')
stopwords.add('U')
stopwords.add('much')
stopwords.add('Dem')
stopwords.add('win')
stopwords.add('know')
stopwords.add('didn')
stopwords.add('time')
stopwords.add('report')
stopwords.add('Dems')
stopwords.add('Left')
stopwords.add('must')
stopwords.add('big')
stopwords.add('said')
stopwords.add('now')
stopwords.add('got')
stopwords.add('one')
stopwords.add('far')
stopwords.add('will')
stopwords.add('totally')
stopwords.add('job')
stopwords.add('go')
stopwords.add('done')
stopwords.add('show')
stopwords.add('want')
stopwords.add('way')
stopwords.add('see')
stopwords.add('allowed')
stopwords.add('won')
stopwords.remove('need')
stopwords.add('made')
stopwords.add('look')
stopwords.add('called')
stopwords.add('us')
stopwords.add('lost')
```

```python
mask = np.array(Image.open("trump_mask.png"))
colors = ["#BF0A30", "#002868"]
cmap = LinearSegmentedColormap.from_list("mycmap", colors)
wc = WordCloud(width = 600, height = 400, random_state=1, background_color='white', max_words=1000, colormap=cmap , mask=mask, collocations=True, stopwords = stopwords).generate(text)
plt.figure(figsize=[15,15])
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
